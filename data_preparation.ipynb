```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import numpy as np
```

```python

df_main=pd.read_csv('student.csv')
```

```python
df_main
```

```python
df_copy= df_main.copy()
```

```python
df_copy['ชื่อ'] = df_copy['ชื่อ'].fillna('ไม่ระบุ')


df_copy
#จัดการข้อมูลส่วนชื่อและเพศที่มีค่าว่างให้เป็นไม่ระบุ
```

```python
df_copy['เพศ'] = df_copy['เพศ'].replace({
    'หญิง': 'ผู้หญิง',
    'ผู้หญิง': 'ผู้หญิง',
    'ญ': 'ผู้หญิง',
    'F': 'ผู้หญิง',
    'Female': 'ผู้หญิง',
    'M': 'ผู้ชาย',
    'Male': 'ผู้ชาย',
    'ชาย': 'ผู้ชาย',
    'ผู้ชาย': 'ผู้ชาย',
    'ช': 'ผู้ชาย'
})
df_copy['คณะ']=df_copy['คณะ'].replace({
    'วิทยาศาสตร์':'วิทยาศาสตร์',
    'คณะวิทยาศาสตร์':'วิทยาศาสตร์',
    'วิทย์':'วิทยาศาสตร์',
    'Science':'วิทยาศาสตร์',
    'Sci':'วิทยาศาสตร์',
    'วิศวะ':'วิศวกรรมศาสตร์',
    'Engineering':'วิศวกรรมศาสตร์',
    'วิศวกรรม':'วิศวกรรมศาสตร์',
    'วิศวกรรมศาสตร':'วิศวกรรมศาสตร์',
    'วิศวฯ':'วิศวกรรมศาสตร์',
    'ศิลปศาสตร์':'ศิลปศาสตร์',
    'Arts':'ศิลปศาสตร์',
    'ศิลป์':'ศิลปศาสตร์',
    'Liberal Arts':'ศิลปศาสตร์',
})

df_copy
#ใช้จัดคำที่มันมีความหมายเหมือนกันแต่ใช้คนละคำ
```

```python
df_copy['คณะ'].fillna(df_copy['คณะ'].mode()[0], inplace=True)
df_copy['เพศ'].fillna(df_copy['เพศ'].mode()[0], inplace=True)
df_copy['ส่วนสูง_cm'].fillna(df_copy['ส่วนสูง_cm'].mode()[0], inplace=True)
df_copy['น้ำหนัก_kg'].fillna(df_copy['น้ำหนัก_kg'].mode()[0], inplace=True)
df_copy

#เติมด้วยค่าที่มีมากที่สุดลงไปในช่องคณะ เพศ ส่วนสูง น้ำหนัก ที่เป็นค่าว่าง
```

```python
mean_gpa = df_copy['GPA'].mean()
df_copy.loc[df_copy['GPA'] < 0, 'GPA'] = np.nan
df_copy.loc[df_copy['GPA'] > 4, 'GPA'] = 4.0
df_copy['GPA'].fillna(mean_gpa, inplace=True)
df_copy
#เราใช้จัดการด้านเกรด GPA คอยเช็คคนที่ GPA มากเกินหรือไม่สมเหตุสมผล
```

```python
mean_19_23 = df_copy.loc[(df_copy['อายุ'] >= 19) & (df_copy['อายุ'] <= 23), 'อายุ'].mean()
condition = (df_copy['อายุ'] > 23) | (df_copy['อายุ'].isna()) | (df_copy['อายุ'] == "")
df_copy.loc[condition, 'อายุ'] = mean_19_23
df_copy
#ใช้จัดการค่าอายุเติมมค่าอายุที่ว่าง
```

```python
maxhours = df_copy['ชั่วโมงเรียน_ต่อสัปดาห์'].max()
minhours = df_copy['ชั่วโมงเรียน_ต่อสัปดาห์'].min()
print(f"ชั่วโมงเรียน_ต่อสัปดาห์: สูงสุด = {maxhours}, ต่ำสุด = {minhours}")
meanHR = df_copy['ชั่วโมงเรียน_ต่อสัปดาห์'].mean()
```

```python
df_copy['ชั่วโมงเรียน_ต่อสัปดาห์'].fillna(meanHR, inplace=True)
df_copy
#เติมค่าช่วงโมงเรียนต่อสัปดาที่เป็นค่าว่าง
```

```python
df_copy['BMI'] = df_copy['น้ำหนัก_kg'] / (df_copy['ส่วนสูง_cm']/100)**2
df_copy['สุขภาพ'] = pd.cut(df_copy['BMI'],
                                   bins=[0, 18.5, 25, 30, 100],
                                   labels=['ผอม', 'ปกติ', 'อ้วน', 'อ้วนมาก'])

print(df_copy[['ชื่อ', 'BMI', 'สุขภาพ']])
#การรวมฟีเจอเพื่อตัวสุขภาพและวิเคราะห์ว่าคนไหนเป็นยังไง
```

```python
reg_data=df_copy.copy()
# ใช้ข้อมูลวันที่ลงทะเบียนเรียน
```

```python
reg_data['วันที่ลงทะเบียน'] = pd.to_datetime(reg_data['วันที่ลงทะเบียน'])
```

```python
count_reg = reg_data.groupby(reg_data['วันที่ลงทะเบียน'].dt.date).size().reset_index(name='จำนวนนศ')

print("แต่ละวันที่นักศึกษาได้ลงทะเบียนเรียน")
print(count_reg)
#ระบุจำนวนนศที่ลงทะเบียนในวันนั้นๆๆๆ
```
